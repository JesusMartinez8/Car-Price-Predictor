Make the unreadable data into data that is trainable

```python
data_clean['Drive'].unique()
```

```python
data_clean['Drive'] = data_clean['Drive'].map({'Front Wheel Drive':0, 'Rear Wheel Drive':1, 'All Wheel Drive': 2})
```

Make sure the data is readable for all aspects that have combined (object and int) variables.

```python
numeric_pattern = r'(\d+\.\d+|\d+)'
data_clean['Acceleration(sec)'] = data_clean['Acceleration'].str.extract(numeric_pattern, expand=False)
data_clean['Acceleration(sec)'] = pd.to_numeric(data_clean['Acceleration(sec)'], errors='coerce')
data_clean['TopSpeed(km/h)'] = data_clean['TopSpeed'].str.extract(numeric_pattern, expand=False)
data_clean['TopSpeed(km/h)'] = pd.to_numeric(data_clean['TopSpeed(km/h)'], errors='coerce')
data_clean['Range(km)'] = data_clean['Range'].str.extract(numeric_pattern, expand=False)
data_clean['Range(km)'] = pd.to_numeric(data_clean['Range(km)'], errors='coerce')
data_clean['Efficiency(Wh/km)'] = data_clean['Efficiency'].str.extract(numeric_pattern, expand=False)
data_clean['Efficiency(Wh/km)'] = pd.to_numeric(data_clean['Efficiency(Wh/km)'], errors='coerce')
data_clean['FastChargeSpeed(km/h)'] = data_clean['FastChargeSpeed'].str.extract(numeric_pattern, expand=False)
data_clean['FastChargeSpeed(km/h)'] = pd.to_numeric(data_clean['FastChargeSpeed(km/h)'], errors='coerce')
data_clean['Price(UK)'] = data_clean['PriceinUK'].str.extract(numeric_pattern, expand=False)
data_clean['Price(UK)'] = pd.to_numeric(data_clean['Price(UK)'], errors='coerce')

data_clean.drop('Acceleration',axis = 1,inplace = True)
data_clean.drop('TopSpeed',axis = 1,inplace = True)
data_clean.drop('Range',axis = 1,inplace = True)
data_clean.drop('Efficiency',axis = 1,inplace = True)
data_clean.drop('FastChargeSpeed',axis = 1,inplace = True)
data_clean.drop('PriceinUK',axis = 1,inplace = True)
```

```python
data_clean = data_clean[~(data_clean['Price(UK)']>125)]
```

```python
data_clean.shape
```

```python
X = data_clean.drop(['Name','Price(UK)'],axis = 1)
Y = data_clean['Price(UK)']
```

```python
Y
```

Split into training set and test set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size= 0.20,random_state=42)
```

Import the models

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

Train the models

```python
lr = LinearRegression()
lr.fit(X_train,Y_train)

rf = RandomForestRegressor()
rf.fit(X_train,Y_train)

xgb = GradientBoostingRegressor()
xgb.fit(X_train,Y_train)

xg = XGBRegressor()
xg.fit(X_train, Y_train)
```

Prediction on Test Data

```python
y_pred1 = lr.predict(X_test)
y_pred2 = rf.predict(X_test)
y_pred3 = xgb.predict(X_test)
y_pred4 = xg.predict(X_test)
```

Evaluate Algorithm

```python
from sklearn import metrics
```

```python
score1 = metrics.r2_score(Y_test,y_pred1)
score2 = metrics.r2_score(Y_test,y_pred2)
score3 = metrics.r2_score(Y_test,y_pred3)
score4 = metrics.r2_score(Y_test,y_pred4)
```

```python
print(score1,score2,score3,score4)
```

```python
final_data = pd.DataFrame({'Models':['LR','RF','GBR','XG'], "R2_SCORE":[score1,score2,score3,score4]})
```

```python
final_data
```

```python
sns.barplot(x =final_data['Models'],y= final_data ['R2_SCORE'])
```
